Example 4: A function that returns the double of a number if it is evena nd triple of a number if it is odd.

```python
def double_triple(number):
    if number % 2 ==0:
        return 2* number
    else:
        return 3* number
print(double_triple(4))
print(double_triple(5))

list_1 = [double_triple(n)for n in range(1,6)]
print(list_1)
```

* args = allows you to pass multiple non-key arguments - tuples
* kwargs = allows you to pass multiple keyword arguments - dictionary

```python
def shipping_label(*args,**kwargs):
    for arg in args:
        print(arg, end=" ")
    print()

    print(f"{kwargs.get('street')},{kwargs.get('apt')}")
    print(f"{kwargs.get('city')} {kwargs.get('state')},{kwargs.get('zip')}")
    
shipping_label("Dr.","Spongebob","Squarepants","III",
               street = "123 Fake St.",
               apt ="100",
               city = "Detroit",
               state = "MI",
                zip ="54321" )
```

Recursion - a function that calls itself from within


```python
# Iterative version
def walk(steps):
    for step in range(1, steps + 1):
        print(f"You take step # {step}")
walk(100)
```

```python
# Recursive 
def walk(steps):
    if steps == 0:
        return
    walk(steps - 1)
    print(f"You take step # {steps}")

walk(1000)
```

```python
# Iterative
def factorial(x):
    result = 1
    if x > 0:
        for y in range(1,x+1):
            result = result * y
        return result
    
# Recursive
def factorial(x):
    if x ==1:
        return 1
    else:
        return x * factorial(x -1)
print(factorial(10))
```

lambda Function
- A small anonymous function for a one time use (throw away function)
Helps keep the namespace clean and is useful with higher order functions.
'sort()','map()','filter()','reduce()'

```python
double = lambda x: x * 2
add = lambda x,y: x + y
print(add(2,3))
```

```python
max_value = lambda x,y: x if x > y else y
min_value = lambda x,y: x if x < y else y
print(min_value(4,5))

full_name = lambda first,last: first + " " + last
print(full_name("Susmi", "Kharel"))

is_even = lambda x: x % 2 == 0
print(is_even(5))

age_check = lambda age: True if age >= 18 else False
print(age_check(21))
```

Quiz 0601 - largest number
Write a python function to find the largest out of 3 numbers. You should use comparison operator to find out the maximum of 3 numbers.

```python
def largest_number(x,y,z):
    max_num = x
    if y > max_num:
        max_num = y
    if z > max_num:
        max_num = z

    return max_num
print(largest_number(1,2,3))
```

```python
def character_count(s):
    upper = 0
    lower = 0
    spaces= 0
    for char in s:
        if char.isupper():
            upper += 1
        elif char.islower():
            lower += 1
        elif char == ' ':
            spaces+= 1
    return (upper,lower,spaces) 
user = input("What is your full name: ")
result = character_count(user)
print(result)
```
